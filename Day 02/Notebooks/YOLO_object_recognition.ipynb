### Selezione dell'immagine

Incolla l'url fra i doppi apici (es. `ur"immagine da scaricare"`) e gira il codice nella cella e in quella successiva


```
url = "https://www.ilmiogattoeleggenda.it/wp-content/uploads/2019/05/Gatto-che-guarda-la-televisione.jpg"
```

```
img = yolo.image_from_url(url)
yolo.view_image(img)
```

### Riconoscimento degli oggetti e visualizzazione dell'immagine

```
img_out = yolo.recognize(img, file_classes="yolov3.txt", file_weights="../yolov3.weights", file_config="yolov3.cfg")
yolo.view_image(img_out)
```
